```python
import numpy as np
import brownian
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nbsteps = 100
trials = 1000

distances = []
for _ in range(trials):
    walker = brownian.BrownianMotioner()
    rms = [0.0] + [np.sqrt(x*x+y*y) for x, y in walker.walk_for_nbsteps(nbsteps)]
    distances.append(np.array(rms))
    
distances = np.array(distances)
```

```python
average_distances = np.mean(distances, axis=0)
```

```python
plt.plot(np.arange(101), average_distances)
```

```python
plt.plot(np.arange(101), np.square(average_distances))
```
